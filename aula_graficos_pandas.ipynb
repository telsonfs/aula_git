### Variaveis Quantitativas 

São aquelas que nós vamos realizar operações matemáticas.

Ex: Preço de um produto, peso, altura...

Obs: Cuidado com os outliers.

## Visualizações das variáveis quantitativas

Boxplot: Localiza possíveis erros na base e indica se a média vai enviesada.


```python
data = np.random.randn(25, 4)
df = pd.DataFrame(data, columns=list('ABCD'))
ax = df.plot.box()
```

Histograma

```python
df = pd.DataFrame(
    np.random.randint(1, 7, 6000),
    columns = ['one'])
df['two'] = df['one'] + np.random.randint(1, 7, 6000)
ax = df.plot.hist(bins=12, alpha=0.5)
```

Gráfico de linha

```python
df = pd.DataFrame({
   'pig': [20, 18, 489, 675, 1776],
   'horse': [4, 25, 281, 600, 1900]
   }, index=[1990, 1997, 2003, 2009, 2014])
lines = df.plot.line()
```

Correlação

```python
df.corr()
```

### Variaveis Qualitativas

São aqueles que dão uma característica ao dado.

Ex: nome, genero, cidade, estado...

```python
df = pd.DataFrame({'lab': ['A', 'B', 'C'], 'val': [10, 30, 20]})
ax = df.plot.barh(x='lab', y='val')
```

```python
speed = [0.1, 17.5, 40, 48, 52, 69, 88]
lifespan = [2, 8, 70, 1.5, 25, 12, 28]
index = ['snail', 'pig', 'elephant',
         'rabbit', 'giraffe', 'coyote', 'horse']
df = pd.DataFrame({'speed': speed,
                   'lifespan': lifespan}, index=index)
ax = df.plot.barh()
```
